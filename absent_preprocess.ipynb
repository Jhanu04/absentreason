# Dropping id because we cant infer anything 

```python
data = data.drop(['ID'], axis=1)
```

# There are actually a lot of reason for the worker's absence now we are grouping similar reasons into 4 common reason_types

```python
reason_columns = pd.get_dummies(data['Reason for Absence'], drop_first=True)
reason_type_1 = (reason_columns.loc[:, 1:14].max(axis=1) > 0).astype(int)
reason_type_2 = (reason_columns.loc[:, 15:17].max(axis=1) > 0).astype(int)
reason_type_3 = (reason_columns.loc[:, 18:21].max(axis=1) > 0).astype(int)
reason_type_4 = (reason_columns.loc[:, 22:28].max(axis=1) > 0).astype(int)
data = data.drop(['Reason for Absence'], axis=1)
reason_type_1.name = 'Reason_Type_1'
reason_type_2.name = 'Reason_Type_2'
reason_type_3.name = 'Reason_Type_3'
reason_type_4.name = 'Reason_Type_4'
data = pd.concat([data, reason_type_1, reason_type_2, reason_type_3, reason_type_4], axis=1)

```

# extracting date,month and day of the week from given date column

```python
data['Date'] = pd.to_datetime(data['Date'], format='%d/%m/%Y')
data['Month'] = data['Date'].apply(lambda x: x.month)
data['Day of the Week'] = data['Date'].apply(lambda x: x.weekday())
data = data.drop(['Date'], axis=1)
```

# 1 refers to highschool 2-4 represents graduate,high graduate so labeling graduate and above as 1 

```python
data['Education'] = data['Education'].map({1: 0, 2: 1, 3: 1, 4: 1})
```
